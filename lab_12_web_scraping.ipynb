```python
import requests

# Fetch a web page
r = requests.get("https://www.softr.io/blog/tools-to-extract-data-from-website")
print(r.text)
```

```python
from bs4 import BeautifulSoup
```

```python

# Remove HTML tags using Beautiful Soup library
soup = BeautifulSoup(r.text, "html5lib")
print(soup.get_text())
```

```python
#mw-content-container
# Find all articles
summaries = soup.find_all("div", class_="ql-editor")
summaries[0]
```
